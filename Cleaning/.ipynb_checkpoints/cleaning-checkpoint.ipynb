```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../Data/cleaned_data.csv')
df.head(10)
```

```python
def sort_desc_uk(x):
    if 'yD' in x:
        return x.split('yD')[0] + 'y'
    elif 'yM' in x:
        return x.split('yM')[0] + 'y'
    
def sort_desc_dk(x):
    if 'yD' in x:
        return 'D' + x.split('yD')[1]
    elif 'yM' in x:
        return 'M' + x.split('yM')[1]

df['description_dk'] = df['description'].apply(sort_desc_dk)
df['description_uk'] = df['description'].apply(sort_desc_uk)
df['wind_direction'] = df['wind_direction'].apply(lambda x: x.split('ind')[1])
df['day_temp'] = df['day_temp'].apply(lambda x: int(re.findall(r'-?\d+', x)[0]))
df['night_temp'] = df['night_temp'].apply(lambda x: int(re.findall(r'-?\d+', x)[0]))
```

```python
df.head(10)
```

```python
df_cleaned = df[['date', 'weekday', 'day_temp', 
                 'night_temp', 'rain_%', 'wind_strength', 
                 'wind_direction', 'description_dk', 'description_uk']].copy()
df_cleaned.head()
```

```python
df_cleaned.to_csv('../Data_Cleaned/cleaned_data.csv', index = False)
```
