```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rc("axes", labelsize=18)
plt.rc("xtick", labelsize=16, top=True, direction="in")
plt.rc("ytick", labelsize=16, right=True, direction="in")
plt.rc("axes", titlesize=18)
plt.rc("legend", fontsize=16)
plt.rc("figure", figsize=(7, 5), titlesize=18)
```

```python
df = pd.read_csv('../Data_Cleaned/cleaned_data.csv')
```

```python
df.head()
```

```python
job_vs_salary = pd.DataFrame({'Salary_mean': [], 'Jop_type': []})
job_vs_salary.loc[0] = [df.loc[df['Job_type'] == 'scientist']['Salary_mean'].mean(), 'scientist'].copy()
job_vs_salary.loc[1] = [df.loc[df['Job_type'] == 'analyst']['Salary_mean'].mean(), 'analyst'].copy()

plt.grid(linestyle = '-', linewidth = 1, axis = 'y', alpha = 0.5)
sns.barplot(data = job_vs_salary, x = 'Jop_type', y = 'Salary_mean')
plt.title('Mean salary of data analysts and scientists')
plt.show()
```

```python
plt.grid(linestyle = '-', linewidth = 1, alpha = 0.5)
sns.scatterplot(data = df, x = 'Rating', y = 'Salary_mean')
plt.title('Correlation between rating and mean salary')
plt.show()
```

```python
sns.kdeplot(data = df, x = 'Salary_mean', hue = 'Job_type', multiple = "stack")
plt.grid(linestyle = '-', linewidth = 1, alpha = 0.5)
plt.title('Distribution of mean salary')
plt.show()
```

```python
sns.kdeplot(data = df, x = 'Rating', hue = 'Job_type', multiple = "stack")
plt.grid(linestyle = '-', linewidth = 1, alpha = 0.5)
plt.title('Distribution of rating for different job titles')
plt.show()
```

```python
rating_vs_size = pd.DataFrame({})
rating_vs_size['Rating'] = df['Rating'].copy()
rating_vs_size['Size'] = ['Big' if data <= 1000 else 'Small' for data in df['Size_mean']]

sns.kdeplot(data = rating_vs_size, x = 'Rating', hue = 'Size', multiple = "stack")
plt.grid(linestyle = '-', linewidth = 1, alpha = 0.5)
plt.title('Distribution of rating for different company sizes')
plt.show()
```
